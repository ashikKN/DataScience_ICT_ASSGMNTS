## **Diabetes Model**

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('/content/diabetes_model.csv')
df.head()
````

**Univariate Analysis:**

````python
plt.hist(df['Glucose'],bins=10,color='lightBlue',edgecolor='black')
plt.xlabel('Glucose')
plt.ylabel('Range')
plt.title('Histogram of Glucose Level')
plt.show()
````

Presence of High Values: Several bins contain high values, indicating some individuals have glucose levels significantly above the normal range.

The distribution of blood pressure is skewed to the right, meaning there are more individuals with lower blood pressure values compared to higher ones.

Bivariate Analysis

````python
sns.scatterplot(x='Age',y='BloodPressure',data=df,color='green')
plt.show()
````

Aging and Blood Pressure:
The positive correlation suggests that aging is a factor associated with increased blood pressure.

The data points tend to cluster in certain areas, especially around the middle age range (30-60).

Outliers: A few data points, particularly at the higher age range, seem to deviate significantly from the general trend

Multivariate Analysis

````python
# pairplot
sns.pairplot(df[['Age','Glucose','Insulin']])
plt.show()
````

Age and Glucose: The weak positive correlation between Age and Glucose might indicate that older individuals are more likely to have higher glucose levels, possibly due to factors like decreased physical activity or metabolic changes associated with aging.

Glucose and Insulin: The strong positive correlation between Glucose and Insulin is consistent with the physiological role of insulin in regulating blood sugar levels. When glucose levels rise, the body produces more insulin to bring them back down.

Insulin Distribution: The highly skewed distribution of Insulin suggests that most individuals have relatively low insulin levels, while a small number have significantly higher levels. This could be due to factors like insulin resistance, diabetes, or other underlying health conditions.
